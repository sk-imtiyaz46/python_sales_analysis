# DATA VISUALIZATION / EXPLORATORY DATA ANALYSIS

## gender

```python
#creating a bar chart on gender basis
br=sbn.countplot(x='Gender', data=df)
for bars in br.containers:
    br.bar_label(bars)               #hue="Gender" for multi colour bars
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

br2=sbn.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, hue='Gender')
for bars in br2.containers:
    br2.bar_label(bars)
```

From The above graphs we can see most of the buyers are female compare to male

```python
br3=sbn.countplot(x="Age Group",data=df,hue='Gender')
for bars in br3.containers:
    br3.bar_label(bars)
```

# Age

```python
#Total Ammount vs Age group
sales_age=df.groupby("Age Group", as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
br4=sbn.barplot(x='Age Group',y='Amount', data=sales_age)
for bars in br4.containers:
    br4.bar_label(bars)
```

From the above Graph we can clearly see that most of the buyers are from age group 26-35

# State

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sbn.set(rc={'figure.figsize':(15,5)})
sbn.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sbn.set(rc={'figure.figsize':(15,5)})
sbn.barplot(data = sales_state, x = 'State',y= 'Amount')
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*

# PRODUCT CATAGORY

```python
sbn.set(rc={'figure.figsize':(20,5)})
ax = sbn.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```
